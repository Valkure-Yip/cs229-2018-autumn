# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Loading the iris dataset
iris = load_iris()
X = iris.data

# Creating a PCA object and fitting the data
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X)

# Plotting the transformed data
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=iris.target, s=50, cmap='viridis')
plt.title('Principal Component Analysis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

In this example, we again load the iris dataset using scikit-learn's load_iris() function, and extract the feature data X from the Bunch object. We then create a PCA object with n_components set to 2, indicating that we want to reduce the dimensionality of the data to 2 principal components. We fit the PCA model to the data using the fit_transform() method of the PCA object, which returns the transformed data in a new array X_transformed.

Finally, we plot the transformed data using a scatter plot, where the x and y coordinates are the values of the first and second principal component, respectively. We color each point according to its corresponding target value (which indicates the species of iris), and add a title and axis labels to the plot.

The output of this program will be a scatter plot showing the transformed data:

In this plot, we can see that the transformed data is now represented in a 2-dimensional space, with the x-axis representing the values of the first principal component (PC1) and the y-axis representing the values of the second principal component (PC2). The points are colored according to the species of iris to which they belong, and we can see that there is some separation between the different species, indicating that the PCA has captured some meaningful structure in the data.
