# Multivariate Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the parameters of the multivariate Gaussian distribution
mu = np.array([0, 0])
sigma = np.array([[1, 0], [0, 1]])

# Create a meshgrid of x and y values
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y

# Evaluate the multivariate Gaussian distribution at each point on the grid
rv = multivariate_normal(mu, sigma)
z = rv.pdf(pos)

# Plot the distribution
plt.contourf(x, y, z)
plt.colorbar()
plt.show()

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the distribution
ax.plot_surface(x, y, z, cmap='viridis')

# Set the axis labels and adjust the viewing angle
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=30, azim=120)

# Show the plot
plt.show()

```
