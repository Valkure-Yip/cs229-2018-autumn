```python
# logistic loss function plot

import matplotlib.pyplot as plt
import numpy as np

# define the log loss function
def log_loss(y_true, y_pred):
    return - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# create a range of values for y_pred
y_pred_range = np.linspace(0.01, 0.99, 100)

# set the value of y_true to 1
y_true = 1

# compute the log loss for each value of y_pred
loss_values = [log_loss(y_true, y_pred) for y_pred in y_pred_range]

# plot the log loss function
plt.plot(y_pred_range, loss_values)
plt.xlabel('y_pred')
plt.ylabel('Log Loss')
plt.title('Log Loss Function')
plt.show()

```
